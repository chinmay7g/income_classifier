```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\Administrator\Desktop\Python Test Data\NPTEL\income(1).csv')
```

```python
sns.countplot(y='JobType', data = data, hue='SalStat')
```

```python
sns.countplot(y='EdType', hue= 'SalStat', data = data)
```

```python
#hence from here we know that people receiving Masters, Doctorate, Prof-School have the ability to earn more than $50,000
#also we infer that only a minimum amount of people are earning more than $50,000 as opposed to the mass majority.
```

```python
sns.countplot(y= 'occupation', hue= 'SalStat', data = data)
```

```python
#we infer executives holding managerial jobs earn almost similarly to the $50,000 boundary
```

```python
sns.distplot(data['capitalgain'], kde= False, bins=10)
```

```python
#shows that majority of the people have capitl gains from investment nearer to zero
```

```python
sns.boxplot(y='hoursperweek', x='SalStat', data= data)
```
